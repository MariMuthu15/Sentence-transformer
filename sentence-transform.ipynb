```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
source_strings = [
    "Show me all Python developers.",
        "List all IoT developers from the 2023 batch.",
        "I want Python students certified in Python.",
        "Give me any 5 candidates from 2024.",
        "Show me all candidates.",
        "Find male candidates from Chennai.",
        "Candidates with Java certifications.",
        "Students from Coimbatore district with Data Science courses.",
        "Candidates who studied 10th standard and passed.",
        "How many students took Deep Learning and are in 2025 batch?",
        "I want three candidates with AI certifications.",
        "Students from 2024 batch who studied in engineering colleges.",
        "Female students from Tirunelveli with backend skills.",
        "All students who passed 12th standard.",
        "How many candidates studied in government colleges in Madurai?",
        "Get me all male candidates from arts colleges.",
        "Students from 2025 batch who studied in self-financed institutions.",
        "Give me five students who enrolled in Python and certified in it.",
        "Show students from 8th standard in Salem.",
        "Find candidates in Chennai with Python or Data Science skills.",
        "Ten female candidates from 2023 batch.",
        "Students from Namakkal district who studied ITI.",
        "Candidates from 2022 with frontend development skills.",
        "I want all students with C programming certifications.",
        "How many IoT developers passed 12th?",
        "Five candidates in ITI.",
        "List IoT developers from Tamil Nadu.",
        "Show me Python developers in Chennai.",
]
```

```python
query = "please consider me as a fullstack developer"
```

```python
source_embeddings = model.encode(source_strings, convert_to_tensor=True)
query_embedding = model.encode(query, convert_to_tensor=True)

cosine_scores = util.cos_sim(query_embedding, source_embeddings)[0]

results = sorted(
    zip(source_strings, cosine_scores.tolist()),
    key=lambda x: x[1],
    reverse=True
)
accuracy =  max(cosine_scores.tolist())
print('Accuracy: ',accuracy)

if accuracy > 0.7:
    print('Correct Phrase Given calling another api or extenral service')
else:
    raise ValueError('Please give valid input')
# print("Search Results:")
# for text, score in results:
#     print(f"{score:.4f} - {text}")
```
